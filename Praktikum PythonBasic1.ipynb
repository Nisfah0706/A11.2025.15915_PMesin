```python
#Latihan variabel
Darjo = 3
Darno = 5
Waginem = 6

#cetak nama dan jumlah
print("Darjo =", Darjo, ", Darno =", Darno, ", Waginem =", Waginem)

totalApel = Darjo + Darno + Waginem
print("Total apel =", totalApel)

#variabel baru
apelBaru = 2
hasilTambah = totalApel + apelBaru
hasilKali = totalApel * 2
hasilBagi = totalApel / 2

print("Setelah tambah apel:", hasilTambah)
print("Total dikali 2:", hasilKali)
print("Total dibagi 2:", hasilBagi)
```

```python
#Latihan variabel a simple conventer
kilometer = 12.25
mil = 7.38

mil_ke_kilometer = mil * 1.61
kilometer_ke_mil = kilometer * 0.62

print(mil, "mil adalah", round(mil_ke_kilometer, 2), "kilometer")
print(kilometer, "kilometer adalah", round(kilometer_ke_mil, 2), "mil")
```

```python
#Latihan operators and expressions
x = 1
x = float(x)

y = 3 * x**3 - 2 * x**2 + 3 * x - 1

print("y =", y)
```

```python
#Latihan : Simple input and output
a = float(input("Masukkan nilai a: "))
b = float(input("Masukkan nilai b: "))

#hasil
tambah = a + b
kurang = a - b
kali = a * b
bagi = a / b

print("Penjumlahan =", tambah)
print("Pengurangan =", kurang)
print("Perkalian =", kali)
print("Pembagian =", bagi)

print("\nYeahhhh")
```
